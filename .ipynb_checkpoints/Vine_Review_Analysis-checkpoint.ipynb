```python
import pandas as pd
```

```python
# import data
df = pd.read_csv("vine_table.csv")
df
```

```python
# filter by more than 20 total votes
lotta_votes_df = df.query('total_votes >= 20') 
lotta_votes_df
```

```python
# filter again by helpfulness
helpful_votes_df = lotta_votes_df.query('helpful_votes/total_votes >= 0.5')
helpful_votes_df
```

```python
# filter again for reviews left by vine = y
vine_yes_df = helpful_votes_df.query('vine == "Y"')
vine_yes_df
```

```python
# filter again for reviews left by vine = n
vine_no_df = helpful_votes_df.query('vine == "N"')
vine_no_df
```

```python
# find total number of reviews for vine=y & vine=n
total_reviews_yes = pd.Series(vine_yes_df['review_id']).count()
print(f'yes = {total_reviews_yes}')

total_reviews_no = pd.Series(vine_no_df['review_id']).count()
print(f'no = {total_reviews_no}')
```

```python
# find number of 5-star reviews for vine=y & vine=n
five_star_yes = vine_yes_df.query('star_rating == 5.0')
five_star_yes_num = pd.Series(five_star_yes['review_id']).count()
print(f'yes = {five_star_yes_num}')

five_star_no = vine_no_df.query('star_rating == 5.0')
five_star_no_num = pd.Series(five_star_no['review_id']).count()
print(f'no = {five_star_no_num}')
```

```python
# the percentage of 5-star reviews for vine=y(paid) & vine=n(unpaid)
yes_percentage = (five_star_yes_num / total_reviews_yes)
print(f'yes = {yes_percentage:.2%}')

no_percentage = (five_star_no_num / total_reviews_no)
print(f'no = {no_percentage:.2%}')
```
